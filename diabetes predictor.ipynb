```python
# Replace 0 in Pregnancies with median
df['Pregnancies'] = df['Pregnancies'].replace(0, df['Pregnancies'].median())

# Replace 0 in Glucose with mean
df['Glucose'] = df['Glucose'].replace(0, df['Glucose'].mean())

# Replace 0 in BloodPressure with mean
df['BloodPressure'] = df['BloodPressure'].replace(0, df['BloodPressure'].mean())

# Replace 0 in SkinThickness with mean
df['SkinThickness'] = df['SkinThickness'].replace(0, df['SkinThickness'].mean())

# Replace 0 in Insulin with median
df['Insulin'] = df['Insulin'].replace(0, df['Insulin'].median())

# Replace 0 in BMI with mean
df['BMI'] = df['BMI'].replace(0, df['BMI'].mean())

# Replace 0 in DiabetesPedigreeFunction with median
df['DiabetesPedigreeFunction'] = df['DiabetesPedigreeFunction'].replace(0, df['DiabetesPedigreeFunction'].median())

# Replace 0 in Age with median
df['Age'] = df['Age'].replace(0, df['Age'].median())
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler

# Initialize the standard scaler
scaler = StandardScaler()

# Standardize the columns (excluding the 'Outcome' column)
columns_to_standardize = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']
df[columns_to_standardize] = scaler.fit_transform(df[columns_to_standardize])
```

```python
plt.figure(figsize=(15,15))
sns.boxplot(df)
```

```python

columns_to_check = ['Pregnancies', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']

# Perform outlier deletion until all values are within the 0.01 to 0.99 quantile range
for col in columns_to_check:
    lower_bound = df[col].quantile(0.01)
    upper_bound = df[col].quantile(0.99)

    # Delete rows where the column values are outside the 0.01 to 0.99 range
    df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
```

```python
df.head()
```

```python
sns.boxplot(df)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.value_counts()
```

```python
## this is data imbalancing
```




*  undersampling:-we take minority class and decrease the value of it to majority class
*  oversampling:-we take the minority class and increase the value of it to the majority class

*smote:-synthetic data increases the minority and increase sample size   





```python
from imblearn.over_sampling import SMOTE
smote=SMOTE(random_state=42)
x_train,y_train=smote.fit_resample(x_train,y_train)
```

```python
x_train.shape
```

```python
y_train.value_counts()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logistic=LogisticRegression()
```

```python
logistic.fit(x_train,y_train)
```

```python
y_pred=logistic.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred,target_names=['diabetic','non-diabetic']))
```
